```python
import re

year_list = []
year_with_text = []
year_between_list = []
to_remove =[]

re_pattern_year = re.compile('([\d]{4,4})')
re_pattern_year_with_text = re.compile('[a-zA-Z.\s]*([\d]{4,4})\Z')
re_pattern_year_between = re.compile('[\s\w.]*([\d]{4})[-\s\w]*[-/und]+[-\s\w]*([\d]{4})[-\s\w]*')

for pic_key in metadata_dic.iterkeys():
    year_string = metadata_dic[pic_key]['metadata_year']

    if year_string != None or len(year_string) == 0:      
        if len(year_string) == 4 and re_pattern_year.match(year_string):
            # bsp: 1933
            metadata_dic[pic_key]['year_from'] = int(year_string)
            metadata_dic[pic_key]['year_to'] = int(year_string)
            year_list.append(year_string)
            
        elif re_pattern_year_with_text.match(year_string):
            # bsp: 'ca. 1959', 'um 1932', 'wohl 1955', 'vor 1844'
            # keine exakte Zeitangaben
            part = re_pattern_year_with_text.search(year_string).group(1)
            metadata_dic[pic_key]['year_from'] = int(part)
            metadata_dic[pic_key]['year_to'] = int(part)
            year_list.append(year_string)
            
        elif re_pattern_year_between.match(year_string):
            # bsp: '1841/1872', '1841-1872', 'um 1501-1502', '1841 - 1872', '1841-07 - 1842-07', '1671-07-21 - 1672-07-21'
            metadata_dic[pic_key]['year_from'] = int(re_pattern_year_between.search(year_string).group(1))
            metadata_dic[pic_key]['year_to'] = int(re_pattern_year_between.search(year_string).group(2))
            year_between_list.append(year_string)
        
        else:
            # remove entry (invalid)
            to_remove.append(pic_key)
    
#for pic_key in to_remove:
#    del metadata_dic[pic_key]

print "year: " + str(len(year_list))
print "year_between: " + str(len(year_between_list))
print "all: " + str(len(year_between_list) + len(year_list)) + "/" + str(len(metadata_dic))
print "to remove: " + str(len(to_remove))
```
